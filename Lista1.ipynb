```python
#4)  Cheque se um numero é ou não perfeito em uma lista
```

```python
def perfeito(n):
    soma=0
    for val in range(1,n):
        if n % val == 0:
            soma += val
    if soma==n:
        return True
    else:
        return False
        
def exibe():
    n = 100
    
    for val in range(1,n+1):
        if(perfeito(val)):
            print(val)
exibe()
```

```python
#7) Defina uma função python unpackList(x) que a partir de uma lista de listas retorne uma lista em que todos os elementos petençam as listas de entrada. Como por exemplo unpackList([[1,2],[’a’,’z’],[[],’zz’]]) retorna a lista [1,2,’a’,’z’,[],’zz’]
```

```python
def unpacklista():
    guardalista = [[1,2],["a","z"],[[],"zz"]]
    x1,x2,x3 = guardalista
    print(x1+x2+x3)
unpacklista()
```

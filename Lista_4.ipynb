```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2

fig=plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.set_title("title")

ax.set_xlabel("x")

ax.set_ylabel("y")

ax.plot(x,y)
```
